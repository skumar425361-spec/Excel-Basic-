**1.** What is Power BI and why is it used in businesses?

**Answer**: Power BI is a software that:
Connects to different data sources (Excel, SQL, APIs, etc.)
Transforms and cleans data
Creates visual reports (charts, graphs, dashboards)
Shares insights with others in real-time

**Why is Power BI used in businesses?**

Businesses use Power BI for several important reasons:

1. Data Visualization
Converts complex data into easy-to-understand visuals
Helps managers quickly understand trends and patterns

2. Better Decision Making
Provides real-time insights
Helps in making fast and accurate business decisions

3. Data Integration
Combines data from multiple sources into one dashboard
Gives a complete view of business performance

4. Automation
Reduces manual reporting work
Automatically updates dashboards when data changes

5. Cost-Effective
More affordable compared to many other BI tools
Suitable for small and large businesses

6. Collaboration & Sharing
Reports can be shared across teams
Teams can access data anytime, anywhere

 **2.** Name and explain the three main components of Power BI

**Answer**:Three Main Components of Power BI

Power BI is a business analytics tool developed by Microsoft. It has three main components:

1. Power BI Desktop

Power BI Desktop is a Windows-based application used to create reports and visualizations. It allows users to connect to various data sources, clean and transform data, and build interactive dashboards. It is mainly used by data analysts and developers.

2. Power BI Service

Power BI Service is a cloud-based platform used to publish, share, and collaborate on reports. Users can access dashboards from anywhere using a web browser and share insights with their team in real-time.

3. Power BI Mobile

Power BI Mobile is an application available for smartphones and tablets. It allows users to view reports and dashboards on the go, providing real-time updates and easy access to business data anytime, anywhere.

**3.** Explain the Power BI workflow

**Answer**: Power BI is a business analytics tool developed by Microsoft. Its workflow describes the steps used to convert raw data into meaningful insights.

1. Data Collection – Import data from sources like Excel, databases, etc.
2. Data Transformation – Clean and prepare data using Power Query
3. Data Modeling – Create relationships and calculations (DAX)
4. Visualization – Build charts, reports, and dashboards
5. Publishing – Upload reports to Power BI Service
6. Sharing & Refresh – Share with users and update data regularly


 **4.** List any four data cleaning tasks that can be performed in Power Query.

**Answer**: Four Data Cleaning Tasks in Power Query

1. Remove Duplicates – Eliminates repeated rows to ensure data accuracy.
2. Handle Missing Values – Fill, replace, or remove null/blank values.
3. Change Data Types – Convert data into correct formats (e.g., text to number, date format).
4. Filter Data – Remove unwanted rows or keep only relevant data.

 **6.** Define Data View, Report View, and Model View. Explain the purpose of each view

**Answer**: Power BI, developed by Microsoft, provides three important views to work with data effectively.

1. Data View

Data View displays the dataset in a tabular format with rows and columns. It allows users to examine the raw data, check for errors, and create calculated columns using DAX. The main purpose of this view is to understand and validate the data.

2. Report View

Report View is the primary workspace used to create visualizations such as charts, graphs, tables, and dashboards. It helps users present data in an interactive and meaningful way for analysis and decision-making.

3. Model View

Model View shows the relationships between different tables in the dataset. It allows users to create, edit, and manage relationships, ensuring proper data structure and enabling accurate analysis.

**7.** Discuss the different data sources that Power BI supports.

**Answer**: Data Sources Supported by Power BI

Power BI, developed by Microsoft, supports a variety of data sources that help users collect and analyze data from different platforms.

1. File-Based Sources

Power BI can import data from files such as Excel, CSV, text, JSON, and XML. These are commonly used for storing and sharing data in small to medium-sized projects.

2. Database Sources

It connects to databases like SQL Server, MySQL, Oracle, and PostgreSQL. These sources are used for structured data and large-scale enterprise applications.

3. Cloud-Based Sources

Power BI supports cloud services such as Azure SQL Database, SharePoint Online, OneDrive, and Google Analytics. These allow access to data from anywhere and support real-time updates.

4. Web and API Sources

Users can extract data from web pages and REST APIs. This helps in fetching real-time or dynamic data from websites and applications.

5. Big Data Sources

Power BI can connect to big data platforms like Hadoop and Spark, which are used for processing very large and complex datasets.

6. Other Sources

It also supports Power BI datasets, OData feeds, and programming tools like R and Python for advanced analytics.
