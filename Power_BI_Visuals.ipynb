**1.** Create a Card Visual to show Total Sales.

**Answer:**
Open Power BI Desktop
 * Load your dataset.

Go to Report View
* Click on the Report View icon

Create a Measure (Total Sales)
* Go to Modeling tab
* Click New Measure
* Write the formula:

    Total Sales = SUM(Sales[SalesAmount])


Insert Card Visual
* From the Visualizations pane, click on Card Visual.

Add Measure to Card
* Drag and drop the Total Sales measure into the Values field of the card.

Format the Card (Optional)
* Go to Format pane
Change:
* Title (e.g., "Total Sales")
* Font size
* Data label color

**2.** Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the highest sales?

**Answer:**
Open Power BI Desktop
* Load your dataset.
Go to Report View
Insert Donut Chart
* From Visualizations pane, click on Donut Chart
Add Fields
* Drag Ship Mode → Legend
*  Drag Total Sales → Values
Check Aggregation
* Make sure it is set to Sum (Power BI does this automatically)
Format (Optional)
* Turn on Data Labels
* Add Title: "Total Sales by Ship Mode"

The ship mode with the largest portion in the donut chart generates the highest sales
(Usually Standard Class in most datasets)

**3.** Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any noticeable seasonal pattern in sales?

**Answer:**
Open Power BI Desktop
* Load your dataset.

Go to Report View
Insert Line Chart
* From Visualizations pane, select Line Chart

Add Fields
* Drag Order Date → Axis
* Drag SalesAmount (or Sales column) → Values

Convert Date into Month & Year
* Click the dropdown on Order Date (Axis)

Select Date Hierarchy
* Keep Year and Month (remove Day if needed)

Check Aggregation
* Ensure Sales is set to Sum

Format (Optional)
* Title: "Sales Trend Over Time"
* Turn on Data Labels if needed

The line chart shows monthly and yearly sales trends
A seasonal pattern exists, with higher sales during festive/end-of-year months

**5.** Create a Matrix Visual to display Sales by Month and Category and identify the month with the highest sales

**Answer:**  
Open Power BI Desktop
* Load your dataset.

Go to Report View

Insert Matrix Visual
* From Visualizations pane, click on Matrix

Add Fields
* Drag Order Date → Rows
* Click dropdown → select Date Hierarchy → keep Month (and Year if needed)
* Drag Category → Columns
* Drag Sales (or SalesAmount) → Values

Check Aggregation
* Ensure it is set to Sum
Enable Totals
* Turn on Row Grand Total to see total sales per month

Apply Conditional Formatting (Optional but Recommended)
* Go to Values → Conditional Formatting → Background Color
* Highest sales will be highlighted automatically

 Identify the month with the highest sales

* Check the Row Total (Total Sales per Month)
* The month with the highest total value is your answer
* With conditional formatting, it will appear as the darkest/hi

**7.** Create a Scatter Chart to compare Discount vs. Profit by Sub-Category. Are there any outliers where high discounts resulted in negative profit?

**Answer:**
Open Power BI Desktop
* Load your dataset.

Go to Report View

Insert Scatter Chart
* From Visualizations pane, select Scatter Chart

Add Fields
* Drag Discount → X-Axis
* Drag Profit → Y-Axis
* Drag Sub-Category → Legend (or Details)
* Drag Sales (optional) → Size (for bubble size)

Check Aggregation
* Ensure Discount = Average (or Sum as needed)
* Profit = Sum

Format (Optional)
* Title: "Discount vs Profit by Sub-Category"
* Turn on Data Labels if needed

**Outlier Identification**

Yes, outliers usually exist:
* Some sub-categories (like Tables, Bookcases) often show:
 * High discount
* Negative profit

This means heavy discounts are causing losses.

Scatter chart shows relationship between Discount and Profit
Clear negative correlation in some cases
Outliers exist where high discounts lead to losses

**9.** Use a Waterfall Chart to analyze how Profit changes by Category within the countries.

**Answer:**
Open Power BI Desktop
* Load your dataset.

Go to Report View

Insert Waterfall Chart
* From Visualizations pane, click on Waterfall Chart

Add Fields
* Drag Category → Category
* Drag Profit → Y-axis (Values)
* Drag Country → Breakdown

Check Aggregation
* Ensure Profit is set to Sum

Format (Optional)
* Add Title: "Profit Change by Category within Countries"
* Turn on Data Labels

How to Analyze
* The chart shows increase/decrease in profit across categories
* Each bar represents:
 * Increase → Profit gain
 * Decrease → Loss
* With Country breakdown, you can see:
* Which country contributes positive or negative profit in each category

**10.** Create a Dual-Axis Line Chart to show Total Sales and Profit Over Time (by Month).

**Answer:**
Open Power BI Desktop
* Load your dataset.

Go to Report View

Insert Line Chart
* From Visualizations pane, select Line Chart

Add Fields
* Drag Order Date → X-Axis
* Select Date Hierarchy → Month (and Year if needed)
