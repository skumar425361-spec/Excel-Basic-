**1.** Define Data Transformation in ETL and explain why it is important.

**Answer**: Data Transformation is the middle step of the ETL process, where data is processed, cleaned, and modified to meet business requirements and ensure compatibility with the destination system.

**Data Transformation is Important**

* Improves Data Quality

Ensures data is accurate, consistent, and reliable by removing errors and inconsistencies.
* Ensures Data Consistency

Standardizes data from different sources into a uniform format.
* Enables Better Analysis

Clean and structured data helps in accurate reporting and decision-making.
* Supports Data Integration

Makes it possible to combine data from multiple systems effectively.
* Enhances Performance

Optimized and structured data improves query speed and system performance.
* Ensures Business Rules are Applied

Applies organizational rules (e.g., calculating profit, categorizing customers).

**4.** A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when they should be used

**Answer**:  If a dataset has missing values in the “Age” column, here are two common techniques to handle them:

**Mean/Median Imputation**

This technique involves replacing missing values with the mean (average) or median of the available Age values.

**Mean Imputation**: Suitable when the data is normally distributed and does not contain extreme outliers.

**Median Imputation**: Preferred when the data is skewed or contains outliers, as it is less affected by extreme values.

**Usage**:

This method is used when we want to retain all records in the dataset and avoid data loss.

**Deletion Method (Removing Records)**

In this method, rows with missing Age values are removed from the dataset.

**Listwise Deletion**: Entire rows with missing values are deleted.

**Usage:**

This technique is appropriate when the number of missing values is very small (e.g., less than 5–10%) and removing them does not significantly affect the dataset or analysis.

**8.** Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist.

**Answer**: Z-score Standardization (or Standardization) and Min-Max Scaling (or Normalization) are two common techniques to scale numerical data.

When outliers exist, Z-score standardization is generally preferred.

**Reason
Min-Max Scaling**
* Scales data to a fixed range (usually 0 to 1) using:
​
* Problem: If the dataset has outliers (extremely high or low values), the minimum and maximum are affected.
* This causes most normal data points to be compressed into a small range, losing sensitivity.

Example:

Data: [10, 12, 15, 1000] → After Min-Max Scaling:

Most values (10, 12, 15) get scaled between 0–0.005, while the outlier dominates the 1.


**Z-Score Standardization**

* Scales data based on mean (μ) and standard deviation (σ):	​

* Advantage: It centers the data around 0 and scales according to variability, so outliers have less impact on the relative position of other points.

Example:

Data: [10, 12, 15, 1000] → Z-scores give meaningful standardized values for all points, not overly compressing normal values.
