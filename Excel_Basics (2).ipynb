1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.


Ans- (a) Relative - A Relative cell reference change automatically. When formula is copied to another cell.

   (b) Absolute - An Absolute cell reference does not change. When the formula is copied to another cell.

   (c) Mixed - A Mixed cell reference looks either the row or the column, but not both.
