**Theoretical Questions**

**1.** What is Data Loading in Power BI and why is it considered the first step of analysis?   


**Answer:** Data loading is the process of importing data from various sources into Power BI for analysis and reporting.

Power BI allows:
* Direct data loading
* Data transformation before loading

Why is it the First Step of Analysis?

Data loading is considered the first step because:

**a.** No Data = No Analysis
You cannot analyze anything unless the data is available in Power BI.

**b.** Foundation for All Further Steps

After loading, you can:
* Clean data (Power Query)
* Transform data
* Create relationships
* Build reports and dashboards

**c.** Data Quality Starts Here

If incorrect or incomplete data is loaded, the analysis results will also be wrong.

**d.** Supports Multiple Sources

It allows combining data from different sources into one model, which is essential for meaningful insights.

**2.** Explain the difference between “Load” and “Transform Data” in Power BI.

**Answer:**
1. Load

* Meaning: Directly imports data into Power BI without making changes.
* Purpose: Quickly bring data for immediate use.
* Process: Data is loaded as it is into the data model.
* Tool Used: “Load” button in Get Data window.
* Use Case: When data is already clean and structured.

2. Transform Data

* Meaning: Opens the data in Power Query Editor to clean and modify it before loading.
* Purpose: Prepare and improve data quality.
* Process: You can edit data before loading it into Power BI.
* Tool Used: “Transform Data” option (Power Query).
* Use Case: When data has errors, missing values, or needs formatting.

**3.** What is a Fact Table and a Dimension Table? Give examples from the dataset.

**Answer:** **Fact Table**

A Fact Table contains quantitative data (numbers/measures) that can be analyzed. It usually includes:
* Sales amount
* Quantity
* Profit
* Revenue

It also contains foreign keys that link to dimension tables.

 Example:
In a sales dataset:

* Columns: Order ID, Product ID, Customer ID, Sales, Quantity, Profit

**Dimension Table**

A Dimension Table contains descriptive information (text/details) about the facts. It helps in filtering, grouping, and labeling data.

Example:
* Customer Table: Customer ID, Customer Name, City, Region
* Product Table: Product ID, Product Name, Category

**4.** Why is Star Schema preferred over Snowflake Schema in Power BI?

**Answer:**  
**Simpler Structure**

Star Schema: Has one central fact table connected directly to dimension tables.

Snowflake Schema: Has multiple related dimension tables (more complex).

 **Better Performance**

Star schema uses fewer joins, so queries run faster in Power BI.

Snowflake schema requires more joins, which can slow down reports.

**Easy Data Modeling**

Relationships in star schema are straightforward (one-to-many).

Snowflake schema has multiple levels, making it harder to manage.

**Faster Report Creation**

With star schema, creating visuals, measures, and dashboards is quicker.

Snowflake schema increases complexity while building reports.

**5.** Identify and remove duplicate records based on Date, Country, and State.

**Answer :**
* **Open Power Query Editor:** Click on 'Transform Data' in the Home tab of Power BI Desktop.
* **Select Columns:** Hold down the Ctr key and click on the headers of the three specific columns: Date, Country and State.
* **Right-Click:** Once all three columns are highlighted, right-click on any of the selected column headers.
* **Remove Duplicates:** Select the "Remove Duplicates" option from the context menu.


**6.** Identify and replace null values in Vaccination_Status.

**Answer:**
 * **Locate the Column:** In the Power Query Editor window, find the column named Vaccination_Status.

* **Identify Nulls:** You will see cells labeled as null. You can also click the filter arrow in the column header to see if "null" is checked.

* **Replace Values:**

* Right-click on the Vaccination_Status column header.
* Select "Replace Values..." from the menu.

* **Enter Details:**
* Value To Find: Type null (Power Query is case-sensitive, so ensure it matches the data).
* Replace With: Type a relevant placeholder, such as "Unknown", "Not Reported", or "0" (depending on the data type and requirement).

* **Apply:** Click OK and then select Close & Apply from the Home ribbon.
