**Easy Level**

**1.** Explain why publishing and sharing reports is considered more important than just building dashboards in Power BI Desktop.

**Answer:** While Power BI Desktop is where you build the "engine," publishing and sharing are what actually drive the business forward. Here is why sharing is the more critical step:

**Democratizing Data:**
 A dashboard sitting on your local drive helps one person. Publishing it to the Power BI Service allows hundreds of stakeholders to access the same "single source of truth" simultaneously from any device.

**Real-Time Collaboration:** Sharing enables features like automated data refreshes. Without publishing, your data is a static snapshot that requires manual updates; sharing ensures everyone is looking at live, relevant numbers
.

**Security and Governance:** Power BI Desktop doesn't offer robust security. By publishing, you can implement Row-Level Security (RLS), ensuring users only see the data they are authorized to see (e.g., a regional manager only seeing their own region).

**Interactive Insights:** Sharing allows users to set their own data alerts, subscribe to email reports, and leave comments. This turns a static visual into a communication tool where teams can discuss trends and take action.

**Version Control:** Keeping a file on a desktop risks version confusion ("Report_Final_v2_Copy"). The Service acts as a centralized hub where the most recent, verified version is always the one being consumed

**Answer:** Power BI Service is a cloud-based online platform by Microsoft that is used to publish, share, and manage reports and dashboards. It allows users to access reports through a web browser or mobile app and enables collaboration across teams.

1. Power BI Desktop (Development Stage):

Power BI Desktop is mainly used for:

*  Data import and transformation
* Creating data models
* Designing reports and visualizations

2. Power BI Service (Sharing & Consumption Stage):

Power BI Service is used for:

* Publishing reports created in Desktop
* Sharing dashboards with users
* Collaboration and access control
* Automatic data refresh and real-time update

**7** Describe the four workspace roles available in Power BI and explain how access control is managed using these roles.

**Answer:** In Power BI Service, workspaces have four roles that control what users can do. These roles help manage access control and security.

1. **Admin**

* Full control over the workspace
* Can add/remove users and assign roles
* Can edit, delete, and publish content

**Access Control Role:**

Admins manage permissions and ensure only the right people have access.

2. **Member**

* Can create, edit, and publish reports and dashboards
* Can share content and collaborate

**Access Control Role:**

Members actively work on content but have limited control over user permissions compared to Admins.

3. **Contributor**

* Can create and edit content
* Cannot manage workspace access or share with others

**Access Control Role:**

Contributors help in development but have restricted sharing permissions for better security.

4. **Viewer**

* Can only view reports and dashboards
* Cannot edit or share content

**Access Control Role:**

Viewers have read-only access, ensuring data is safe from unwanted changes.

**8.** Why are Power BI Apps considered more secure and scalable than direct sharing?

**Answer:** Explain this from a governance and enterprise usage perspective.
From a governance and enterprise perspective, Power BI Apps are superior to direct sharing because they provide a formal distribution layer that separates development from production

1. **Governance Perspective (Security & Control)**

**Power BI Apps:**

* Content is centrally managed by admins
* Users get read-only access, reducing risk of accidental changes
* Permissions are assigned at app level, not individually
* Ensures consistent and approved data is shared

**Direct Sharing:**

* Access is given to users individually
* Harder to track who has access
* Higher risk of unauthorized sharing or data leakage

2. **Scalability Perspective (Enterprise Usage)**

**Power BI Apps:**

* Designed to share with large audiences (hundreds or thousands of users)
* Easy to update—changes reflect automatically for all users
* Provides a structured and organized experience

**Direct Sharing:**

* Not efficient for large groups
* Requires manual sharing for each user
* Difficult to maintain as the number of users grows

**9.** Explain Row-Level Security (RLS) in Power BI. How does RLS ensure that different users see different data using the same report?

**Answer:** Row-Level Security (RLS) is a Power BI feature that restricts data access at the row level for specific users. It ensures that even though multiple users access the same report, they only see the data they are authorised to view

**How RLS Works:**

*  RLS is defined in Power BI Desktop by creating roles.
* Each role has filters applied to the dataset (e.g., Region = "East").
* Users are assigned to these roles after the report is published.

**How RLS Ensures Different Data for Different Users:**

* When a user opens the same report, Power BI checks their assigned role.
* The system automatically applies the filter conditions linked to that role.
* As a result, each user sees only the data allowed for them, even though the report is the same.

**Example:**
A company has sales data for multiple regions.
Manager A (East) sees only East region data.
Manager B (West) sees only West region data.

**10.** Assume an organization shares reports without using Apps, workspace roles, or RLS.
Discuss any three risks or issues that may arise related to security, governance, or data misuse, and explain
how Power BI best practices help prevent them.


**Answer:** If an organization shares reports in Power BI Service without using Apps, workspace roles, or Row-Level Security (RLS), several risks can arise.

1. **Data Security Risk (Unauthorized Access)**

**Issue:**

Without RLS or proper roles, all users may see complete data, including sensitive information (e.g., financial or personal data).

**Best Practice Solution:**

* Implement Row-Level Security (RLS) in Power BI Desktop
* Assign users to roles so they only see relevant data

2. **Poor Governance (Lack of Control)**

**Issue:**

Without workspace roles, there is no clear control over who can edit, share, or delete reports. This can lead to confusion and errors.

**Best Practice Solution:**

* Use workspace roles (Admin, Member, Contributor, Viewer)
* Restrict permissions based on user responsibility

3. **Data Misuse or Inconsistency**

**Issue:**

Direct sharing without Apps can result in:

* Multiple versions of reports
* Users accessing outdated or unverified data
* Misinterpretation of insights

** Best Practice Solution:**

* Use Power BI Apps to distribute approved and consistent reports
* Ensure users access a single, trusted version of data
