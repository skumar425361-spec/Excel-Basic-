3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?

**Answer:**
Mean ( Average) = 50

Standard deviation = 5

Range 45 to 55

**Understand the range**

45 is 5 less than the mean → one standard deviation below

55 is 5 more than the mean → one standard deviation above

           45 = 50−5 = μ−1σ

          55 = 50+5 = μ+1σ

So the range 45 to 55 = mean ± 1 standard deviation
Normal distribution rule (important)

For a normal (bell-shaped) distribution:

 68% of the data lies within 1 standard deviation of the mean

Answer - 68%

8. How does log transformation help in stabilizing variance and making data more normally distributed?
**Answer:** A log transformation is when you take the logarithm of each data point, usually:

                      y=log⁡(x)
 Common bases: log base 10, natural log (ln), or log base 2.

**How Log Transformation Helps**

**Stabilizes Variance (Reduces Heteroscedasticity)**

* Some datasets have variance that increases with the mean (large values vary more than small values)
* Taking the log compresses large values and spreads small values, making variance more constant across the range

**Makes Data More Normally Distributed**

* Many statistical methods (t-tests, ANOVA, regression) assume normality

* Highly skewed data (right-skewed, long tail) can be transformed

* Log transformation pulls in extreme high values and reduces skewness
